```python
import http.client
import json
import requests
import pandas as pd
import math 
keyword = "Data Science"
location = "Dallas"
page = "1"

key = "5e885af9-75bb-4d99-b61b-29451e56addb"
# host = 'us.jooble.org'
host = f"https://us.jooble.org/api/{key}"
headers = {"Content-type": "application/x-www-form-urlencoded"}
request = '{"keywords": "' + str(keyword) + '", "location": ' + '"' + str(location) +'", "page": ' + '"' + str(page)+ '"' + '}'

resultsPerPage = 20
response = requests.post(host,headers=headers,data=request)
# print(f"Link: {response.url}")
response = response.json()
# print(response)
totalcount = response["totalCount"]
totalPages = totalcount/resultsPerPage

pagecounter = math.ceil(totalPages)

title = []
location = []
snippet = []
salary = []
source = []
jobtype = []
link = []
company = []
updated = []
jobid = []

print(totalPages)

def toappend(tit,loc,sni,sal,sou,jobt,lin,com,upd,jobi):
    title.append(tit)
    location.append(loc)
    snippet.append(sni)
    salary.append(sal)
    source.append(sou)
    jobtype.append(jobt)
    link.append(lin)
    company.append(com)
    updated.append(upd)
    jobid.append(jobi)


def jobparser(data):
    # print(data)
    def tryandcatch(item):
        try:
            output = data[item]
#             print(output)
            return output
        except:
            output = "Missing"
#             print(output)
            return output
    title = tryandcatch("title")
    location = tryandcatch("location")
    snippet = tryandcatch("snippet")
    salary = tryandcatch("salary")
    source = tryandcatch("source")
    jobtype = tryandcatch("type")
    link = tryandcatch("link")
    company = tryandcatch("company")
    updated = tryandcatch("updated")
    jobid = tryandcatch("id")
    datatoappend = [title,location,snippet,salary,source,jobtype,link,company,updated,jobid]
    toappend(*datatoappend)

    
    # print(f"Found title: {title}")

print(f"Total Counts found: {totalcount}")
counter = 1
while counter <= pagecounter:
    page = counter
    print(f"On Page {counter}, out of {pagecounter}")
    request = '{"keywords": "' + str(keyword) + '", "location": ' + '"' + str(location) +'", "page": ' + '"' + str(page)+ '"' + '}'
    response = requests.post(host,headers=headers,data=request)
    response = response.json()
    jobs = response["jobs"]
    for job in jobs:
        jobparser(job)
    pagecounter -= 1
    counter += 1
    if counter > 5:
        break
    


# print(jobs)


frame = [title,location,snippet,salary,source,jobtype,link,company,updated,jobid]
frameofdata = pd.DataFrame(frame)
frameofdata = frameofdata.T
frameofdata = frameofdata.rename(columns={0:"Job Title",
                                         1:"Location",
                                         2:"Snippet",
                                         3:"Salary",
                                         4:"Source",
                                         5:"Job Type",
                                         6:"Link",
                                         7:"Company",
                                         8:"Updated",
                                         9:"JobID"})
frameofdata.to_csv("Joogle_data_set.csv")
frameofdata
```
