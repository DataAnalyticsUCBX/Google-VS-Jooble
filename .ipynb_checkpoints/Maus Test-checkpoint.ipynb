```python
# host = 'data.usajobs.gov'
# userAgent = 'maurop13@gmail.com'
# authKey = 'Zcf2Xf8NWEc5BDbpRN2SCY26hpAEmmgo7J8ykmRqy3E'
# # url = "https://data.usajobs.gov/api/search"
# # url
# headers = {'Host: data.usajobs.gov','User Agent: maurop13@gmail.com','Authorization Key: Zcf2Xf8NWEc5BDbpRN2SCY26hpAEmmgo7J8ykmRqy3E"} 
# https://data.usajobs.gov/api/search?JobCategoryCode=2210
# import requests
# resp = requests.get("https://data.usajobs.gov/api/jobs?CountrySubdivision=California")
# # at this point, download has completed, now convert it
# # to a data structure
# data = resp.json()
# data
# # print("Total jobs:", data['TotalJobs'])
# # jobs = data['JobData']
# # print("First job title:", jobs[0]['JobTitle'])

import requests

headers = {
    'Host': 'data.usajobs.gov',
    'User-Agent': 'maurop13@gmail.com',
    'Authorization-Key': 'Zcf2Xf8NWEc5BDbpRN2SCY26hpAEmmgo7J8ykmRqy3E=',
}

keywords= "Data"
SearchResultCountAll = 5
dfJobTitle = []
dfLocation=[]
dfcity=[]
dflong=[]
dflat=[]
dfremunmin=[]
dfremunmax=[]
dftyperemun=[]
dfpubdate=[]
dfquals=[]
params = {"keywords":keywords,"ResultsPerPage":SearchResultCountAll}

response = requests.get(' https://data.usajobs.gov/api/search', headers=headers,params = params)
response_json = response.json()
response_json
for i in response_json["SearchResult"]["SearchResultItems"]:
    dfJobTitle.append(i["MatchedObjectDescriptor"]["PositionTitle"])
    dfLocation.append(i["MatchedObjectDescriptor"]["PositionLocationDisplay"])
    dfcity.append(i["MatchedObjectDescriptor"]["PositionLocation"][0]["LocationName"])
    dflong.append(i["MatchedObjectDescriptor"]["PositionLocation"][0]["Longitude"])
    dflat.append(i["MatchedObjectDescriptor"]["PositionLocation"][0]["Latitude"])
    dfremunmin.append(i["MatchedObjectDescriptor"]["PositionRemuneration"][0]["MinimumRange"])
    dfremunmax.append(i["MatchedObjectDescriptor"]["PositionRemuneration"][0]["MaximumRange"])
    dftyperemun.append(i["MatchedObjectDescriptor"]["PositionRemuneration"][0]["RateIntervalCode"])
    dfpubdate.append(i["MatchedObjectDescriptor"]["PublicationStartDate"])
    dfquals.append(i["MatchedObjectDescriptor"]["QualificationSummary"])

results_df = pd.DataFrame({"Job title":dfJobTitle,"Location":dfLocation, "City":dfcity, "Longitude":dflong, "Latitude":dflat,"Minimum remuneration":dfremunmin,"Maximum Remuneration":dfremunmax, "Type of Remuneration":dftyperemun, "Publication Date":dfpubdate,"Qualifications":dfquals})
results_df
```
